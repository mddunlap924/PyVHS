## User Inputs

Locate a video and segment that contains the template you want to remove. When I performed this step I played a video in VLC player, found a segment of the video with the blue screen, and noted the time interval it started and ended.

The user inputs below can clip sections of a video to retrieve the template of interest.

````python
# Path to the video containing the template
PATH = (f'/nvme4tb/videos_to_edit/'
        f'video1.mp4')

# Time the template appears (seconds)
TIME = (1.00)

# Path to save the template image
PATH_TEMPLATE = (f'../pyvhs/template_imgs/template001.png')
````

## Import Libraries and Modules

````python
from moviepy.editor import VideoFileClip
import matplotlib.pyplot as plt
from PIL import Image
````

## Extract video segment

````python
# Create video clip
clip = VideoFileClip(PATH, fps_source='fps')
print(f'Clip Duration [secs]: {clip.duration}')
print(f'# of Frames: {clip.reader.nframes}')
````

````python
#Extract the template image
img = clip.get_frame(t=TIME)
img = Image.fromarray(img, 'RGB')
print(f'Img type: {type(img)}')
print(f'Img Shape: {img.size}')

# Display the template image
plt.imshow(img)
plt.title('Template Image - Blue Screen')
plt.show()
````

````python
# Save template image to disk
img.save(PATH_TEMPLATE)
print(f'Template Image Saved at:\n\t{PATH_TEMPLATE}')
````
